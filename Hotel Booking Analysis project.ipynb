```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.describe(include = "object").T
```

```python
df['children'] = df['children'].fillna(0)  #mean = 0.13
df['country'] = df['country'].fillna('Undefined')
df['agent'] = df['agent'].fillna(int(df['agent'].mean())) #mean = 94.693382
```

```python
df.drop(['company'],axis =1)  #82137 null
```

```python
print ('Hotel :', df.hotel.unique())
```

```python
df.hotel.agg(["value_counts"])
```

```python
hotel = ["City Hotel", "Resort Hotel"]
booking = df.hotel.value_counts()
color = ['purple', 'orange']

plt.figure(figsize = (10, 7))
plt.pie(booking, labels = hotel, colors = color, explode = (0, 0.1), textprops = {'color':"w"}, autopct = "%.2f%%",startangle = 90)

plt.legend()

plt.show();
```

```python
df.groupby(["hotel"])["is_canceled"].value_counts()  #.agg(["value_counts"]) print as a table
```
